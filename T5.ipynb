```python

plt.figure(figsize=(10,8))
df['Region'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
#Profit vs Discount
```

```python
from matplotlib import style
style.use('dark_background')
fig,ax=plt.subplots(figsize=(16,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
#Profit vs Quantity
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['red','blue'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each region')
sns.barplot(x='Region',y='Sales',data=df,hue='Segment',order=df['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['yellow','skyblue'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['pink','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

```python
top_states = df['State'].value_counts().nlargest(10)
top_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/loss and Sales')
plt.show()
```

# Observation:

1.Technology and Office Supplies have high profit.

2.Furniture have less Profit.



```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','purple'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```
